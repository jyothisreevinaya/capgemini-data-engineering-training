```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType

schema = StructType([
    StructField("Item_Identifier", StringType(), True),
    StructField("Item_Weight", DoubleType(), True),
    StructField("Item_Fat_Content", StringType(), True),
    StructField("Item_Visibility", DoubleType(), True),
    StructField("Item_Type", StringType(), True),
    StructField("Item_MRP", DoubleType(), True)
])
```

```python
data = [
    ("FDA15", 9.3, "Low Fat", 0.016047301, "Dairy", 249.8092),
    ("DRC01", 5.92, "Regular", 0.019278216, "Soft Drinks", 48.2692),
    ("FDN15", 17.5, "Low Fat", 0.016760075, "Meat", 141.618),
    ("FDX07", 19.2, "Regular", 0.0, "Fruits and Vegetables", 182.095),
    ("NCD19", 8.93, "Low Fat", 0.0, "Household", 53.8614),
    ("FDP36", 10.395, "Regular", 0.0, "Baking Goods", 51.4008),
    ("FDO10", 13.65, "Regular", 0.012741089, "Snack Foods", 57.6588),
    ("FDP10", None, "Low Fat", 0.127469857, "Snack Foods", 107.7622)
]
```

```python
df = spark.createDataFrame(data, schema=schema)
df.show()
```

```python
df.createOrReplaceTempView("items_table")
```

```python
from pyspark.sql.functions import col

df.select(col('Item_Identifier').alias('Item_id')).show()
```

```python
df.filter(col('Item_Fat_Content') == 'Regular').show()
```

```python
df.filter(
    (col('Item_Type') == 'Soft Drinks') & (col('Item_Weight') < 10)
).show()
```

```python
df.withColumnRenamed('Item_Weight', 'Item_wt').display()
```

```python
from pyspark.sql.functions import lit

df.withColumn('flag', lit('new')).display()
```

```python
from pyspark.sql.functions import col

df.withColumn(
    'multiply',
    col('Item_Weight') * col('Item_MRP')
).display()
```

```python
from pyspark.sql.functions import regexp_replace

df.withColumn(
    'Item_Fat_Content',
    regexp_replace(col('Item_Fat_Content'), 'Regular', 'Reg')
).display()
```

```python
from pyspark.sql.functions import col

df.withColumn(
    'Item_Weight',
    col('Item_Weight').cast('int')
).show()
```

```python
from pyspark.sql.functions import col

df.orderBy(
    col('Item_Weight').desc(),
    col('Item_Visibility').desc()
).show()
```

```python
df.limit(5).show()
```

```python
df.drop('Item_Visibility').show()
```

```python
df1 = spark.createDataFrame([
    (1, "John"),
    (2, "Alice")
], ["id", "name"])

df2 = spark.createDataFrame([
    (3, "Bob"),
    (4, "Sam")
], ["id", "name"])

df1.union(df2).show()
```

```python
df1.unionByName(df2).show()
```

```python
from pyspark.sql.functions import col, current_date, date_add, date_sub, datediff, date_format

df.withColumn("formatted", date_format(col("today"), "dd-MM-yyyy"))
df.display()
```

```python
from pyspark.sql.functions import col, current_date, date_add, date_sub, datediff, date_format

df.withColumn("diff", datediff(col("next_week"), col("today")))
df.display()
```

```python
df.dropna(subset=["Item_Weight"]).display()
```

```python
df.fillna({"Item_Weight": 0}).show()
```

```python
from pyspark.sql.functions import split

df.withColumn("split_col", split(col("Item_Type"), " ")).display()
```

```python
df.withColumn("first_word", split(col("Item_Type"), " ")[0]).display()
```

```python
from pyspark.sql.functions import explode

df.withColumn("words", explode(split(col("Item_Type"), " "))).display()
```

```python
from pyspark.sql.functions import when

df.withColumn("veg_flag",
    when(col("Item_Type") == "Meat", "Non-veg")
    .otherwise("Veg")
).display()
```

```python
from pyspark.sql.functions import sum, avg

df.groupBy("Item_Type").agg(sum("Item_MRP")).display()
```

```python
df.groupBy("Item_Type").agg(
    sum("Item_MRP").alias("total_mrp"),
    avg("Item_MRP").alias("avg_mrp")
).show()
```

```python
from pyspark.sql.functions import collect_list

df.groupBy("Item_Type").agg(
    collect_list("Item_Identifier")
).show(truncate=False)
```

```python
df.groupBy("Item_Fat_Content") \
  .pivot("Item_Type") \
  .agg(avg("Item_MRP")) \
  .show()
```

```python
from pyspark.sql.functions import current_date

df.withColumn("today", current_date()).show()
```
