```sql
select max(avg_amount) as max_Avg 
from (select e.department, avg(s.amount) as avg_amount
from employee e join sales as s
on e.emp_id=s.emp_id
group by e.department) t
```

```sql
select department,total_amount from (
  select e.department, sum(s.amount) as total_amount
  from Employee e join sales as s
  on e.emp_id=s.emp_id
  group by e.department
) t  
order by total_amount desc
limit 1;
```

```sql
select emp_name, total_sales from
(select e.emp_id, e.emp_name , sum(s.amount) as total_sales
from employee e join sales s
on e.emp_id=s.emp_id
group by e.emp_id, e.emp_name ) t
order by total_sales desc
limit 3;
```
