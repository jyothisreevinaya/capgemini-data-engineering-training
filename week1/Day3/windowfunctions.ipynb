```sql
SELECT  emp_id,emp_name, department, salary, join_date,
    ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num
FROM employees;
```

```sql
SELECT emp_id,emp_name, department,
    salary,
    join_date,
    ROW_NUMBER() OVER (
        PARTITION BY department 
        ORDER BY salary DESC
    ) AS dept_row_num
FROM employees;
```

```sql
select emp_id, emp_name, join_date,
row_number() over (order by join_date desc) as latest
from employees;
```

```sql
select emp_id, emp_name, join_date,
row_number() over (partition by department order by join_date asc) as latest
from employees;
```

```sql
select emp_id, emp_name, salary,
row_number() over (order by salary asc) as salary_order
from employees;
```

```sql
select emp_id, emp_name,
row_number() over (order by emp_name asc) as alphabetical_order
from employees;
```

```sql
select emp_id, emp_name, salary,
rank() over (order by salary desc) as highest
from employees;
```
