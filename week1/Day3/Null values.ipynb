```sql
SELECT COUNT(*) AS null_manager_count
FROM employees11
WHERE manager_id IS NULL;
```

```sql
SELECT *
FROM orders11
WHERE discount IS NOT NULL;
```

```sql
SELECT COALESCE(salary, 0) AS salary
FROM employees11;
```

```sql
SELECT COALESCE(bonus, 1000) AS bonus
FROM employees11;
```

```sql
SELECT COALESCE(amount, 500) AS order_amount
FROM orders11;
```

```sql
SELECT COALESCE(stock, 0) AS stock
FROM products11;
```

```sql
SELECT COALESCE(salary, bonus) AS earnings
FROM employees11;
```

```sql
SELECT COALESCE(price, 1000) AS price
FROM products11;
```

```sql
SELECT NULLIF(salary, 0) AS salary
FROM employees11;
```

```sql
SELECT amount , NULLIF(discount, 0) AS result
FROM orders11;
```

```sql
SELECT 
    emp_id,
    name,
    COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM Employees11;
```

```sql
SELECT *
FROM Employees11
WHERE salary IS NULL 
  AND bonus IS NULL;
```

```sql
SELECT *
FROM Products11
WHERE price IS NULL 
  AND category IS NOT NULL;
```

```sql
SELECT *
FROM Orders11
WHERE amount IS NULL 
  AND discount IS NULL;
```

```sql
SELECT 
    emp_id,
    name,
    COALESCE(salary, bonus, 1000) AS income
FROM Employees11;
```

```sql
SELECT 
    order_id,
    NULLIF(discount, 0) AS discount
FROM Orders11;
```

```sql
SELECT 
    order_id,
    COALESCE(amount, 0) - COALESCE(discount, 0) AS final_amount
FROM Orders11;
```
